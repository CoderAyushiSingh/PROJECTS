```python
#“The dataset contained multiple categorical features such as gender,
# school type, parental job, and support systems. I applied one-hot encoding
# to convert them into numerical form and dropped the first category to prevent multicollinearity.”

#ONE HOT ENCODING converts categorical data into numerical data so that machine learning could understand it

x = pd.get_dummies(x, drop_first=True)
```

```python
#train test split
from sklearn.model_selection import train_test_split # This function splits the dataset into training and testing parts.Model should learn from one part Model should be tested on unseen data
```

```python
x_train , x_test , y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

#20% data → testing , 80% data → training
#random_state=42 Fixes randomness Ensures same split every time
```

#Question and answers

 Q1: Why do we split the data?

Answer: To evaluate how well the model performs on unseen data and avoid overfitting.

Q2: What happens if we don’t split?

Answer: The model may memorize the data, giving misleading performance metrics.

Why not 90–10 or 70–30?

Answer: 80–20 is a balanced and commonly accepted split that works well for medium-sized datasets.

Q4: What is overfitting?

Answer: When a model performs well on training data but poorly on new data.

```python
#SCALING = FAIRNESS OF FEATURES
#FEATURE SCALING = iska mtlb sare features ko ek level pr lana
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()  #Data ko is form mein laata hai: Mean = 0 Std = 1 Isse: Bade numbers chhote ho jaate hain Chhote numbers bade ho jaate hain
```

```python
#fit → training data ka mean aur std seekhta hai transform → usi formula se data ko scale karta hai Why only train? Kyuki test data future ka data hai. Uska info pehle nahi lena.
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test) #only fit on train to avoid data leakage
```

#QUESTION ANSWERS
1. Why scaling is important?
ANS =  Because many machine learning algorithms are sensitive to feature magnitudes, and scaling prevents features with large values from dominating the model.

2. do every model needs scaling?
no tree based doesn't need.


```python
#Model Training (Linear Regression) I used linear regression as a baseline

from sklearn.linear_model import LinearRegression
lr = LinearRegression()    #“I initialized the Linear Regression model.”
lr.fit(x_train_scaled , y_train)   #“The model learns the relationship between features and the target by minimizing error.”
```

#QUESTION ANSWERS
1. Why did you choose Linear Regression first?

Ans: Because it is simple, interpretable, and serves as a good baseline model.

2. What does fit() do?

Ans:It trains the model by learning coefficients from the training data.

3. What does the model actually learn?

Ans: It learns feature weights that best explain the target variable.

4.Can Linear Regression overfit?

Ans:Yes, especially when there are many features or multicollinearity.

```python
y_pred = lr.predict(x_test_scaled)     #“I used the trained model to predict target values on the test dataset.”

#predict == ye ek function h jo har row ke liye formula lgata h har student ke liye ek predicted score nikalta h
```

#question answers
1. Why do we predict on test data, not training data?

To evaluate how the model performs on unseen data.

2. What is y_pred?

Answer:It contains the predicted values generated by the model.

3. Is prediction same as evaluation?

No. Prediction gives outputs, evaluation measures how good those outputs are.

```python
#EVALUATION  how wrong it is predicting
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred)     #“MAE tells us the average absolute difference between actual and predicted values.”
mse = mean_squared_error(y_test, y_pred)     #“MSE penalizes larger errors more heavily.”
rmse = np.sqrt(mse)                           #“RMSE is preferred because it penalizes large errors and is in the same unit as the target variable.”
r2 = r2_score(y_test, y_pred)                 #“R² score shows how much variance in the target variable is explained by the model.”

print("MAE:" , mae)
print("MSE:", mae)
print("RMSE:", rmse)
print("r2 score:", r2)
```

#Questions answers
1. Which metric is best?

: It depends on the problem, but RMSE is commonly preferred for regression tasks.

2. High R² but high RMSE?

Answer:The model explains variance well but still makes large prediction errors.

3. Low R² always bad?

Answer: Not necessarily. It depends on data complexity and problem domain.

Linear Regression captures linear relationships, while Decision Trees can model non-linear patterns.

2. Why no scaling?

3. Risk of Decision Tree?

```python
#RANDOM FOREST REGRESSOR >> Multiple Decision Trees ka group Har tree: Data ka thoda sa hissa dekhta hai Alag-alag rules seekhta hai Final output:Sab trees ka average prediction
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 100,random_state = 42 )  #estimators=100 trees,random=reproducibility,i used 100 tress to reduce variance and improve generalzation
rf.fit(x_train, y_train)
y_pred_rf = rf.predict(x_test)
```
